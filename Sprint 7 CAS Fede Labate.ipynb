### Ejercicio 1
#### Calculadora de l"칤ndex de massa corporal
* Escribe una funci칩n que calcule el IMC ingresado por el usuario/a, es decir, quien lo ejecute deber치 ingresar estos datos.
Puedes obtener m치s informaci칩n de su c치lculo en:
[칈ndice de masa corporal IMC que es y c칩mo se calcula](https://muysalud.com/salud/indice-de-masa-corporal-imc-que-es-y-como-se-calcula/)
* La funci칩n debe clasificar el resultado en sus respectivas categor칤as.

```python
#Funci칩n para que el usuario introduzca su peso.
def obtener_peso():
    while True:
        try:
            return float(input("Introduzca su peso en kilos (Ejemplo: 65.5): "))
        except ValueError:
            print("Por favor ingrese su peso en n칰meros, si tiene decimales utilice el punto")

#Funci칩n para que el usuario introduzca su estatura.
def obtener_estatura():
    while True:
        try:
            return float(input("Introduzca su estatura en metros (Ejemplo: 1.70): "))
        except ValueError:
            print("Por favor ingrese su estatura en n칰meros, si tiene decimales utilice el punto")

#Funci칩n para calcular el IMC.
def calcular_IMC():
    peso = obtener_peso() #Llamo a la funci칩n y me devuelve un valor de peso.
    estatura = obtener_estatura() #Llamo a la funci칩n y me devuelve un valor de estatura.
    IMC = round(peso / (estatura**2), 1) #Hago la operaci칩n para saber el IMC
    casos_de_IMC(IMC) #Llamo a la funci칩n y me devuelve el resultado que quiero mostrar.

#Funci칩n de mensajes de los 4 tipos de IMC.
def casos_de_IMC(IMC):
    if IMC < 18.5:
        print(f"Tu IMC es {IMC} e indica que est치s bajo de peso")
    elif IMC >= 18.5 and IMC < 24.9:
        print(f"Tu IMC es {IMC} e indica que est치s en un peso normal")
    elif IMC >= 24.9 and IMC < 29.9:
        print(f"Tu IMC es {IMC} e indica que tienes sobrepeso")
    else:
        print(f"Tu IMC es {IMC} e indica que tienes obesidad")

#Funci칩n principal que maneja el flujo del programa.
def main():
    while True:
        calcular_IMC()
        while True:
            reinicio = input("쯈uieres calcular otro IMC? (SI/NO): ").strip().upper()
            if reinicio == "SI":
                break
            elif reinicio == "NO":
                print("Gracias por usar el calculador de IMC")
                return
            else:
                print("Respuesta inv치lida. Por favor, ingresa 'SI'o 'NO'.")

if __name__ == "__main__":
    main()

#Cuando colocas if __name__ == "__main__": en tu archivo de Python, le est치s diciendo a Python que ejecute el bloque de c칩digo que sigue (en este caso, main()) 
#solo si el archivo se est치 ejecutando como el script principal y no si se est치 importando como un m칩dulo en otro archivo.
```

### Ejercicio 2
#### Convertidor de temperaturas.

Existen diversas unidades de temperatura utilizadas en distintos contextos y regiones. 
Las m치s comunes son:
* Celsius (춿C)
* Fahrenheit (춿F)
* Kelvin (K).

Tambi칠n existen otras unidades como:
* Rankine (춿Ra) 
* R칠aumur (춿Re). 

Selecciona al menos 2 conversores, de modo que al introducir una temperatura devuelva, como m칤nimo, dos conversiones.

```python
#Funci칩n para que el usuario elija el tipo de conversor.
def elegir_conversor():
    print("Ingrese el tipo de conversor que desea utilizar:")
    while True:
        conversor = input("1-Celsius (춿C) a Fahrenheit (춿F) y Kelvin (K)\n"
                          "2-Fahrenheit (춿F) a Celsius (춿C) y Kelvin (K)\n")
        if conversor in ["1", "2"]:
            return int(conversor)
        else:
            print("Entrada incorrecta. Por favor, ingrese 1 o 2.")

#Funci칩n para obtener la temperatura del usuario.
def obtener_temperatura(tipo):
    while True:
        temp_str = input(f"Ingrese la temperatura en grados {tipo}: ")
        try:
            temp = float(temp_str)
            return temp
        except ValueError:
            print("Por favor, ingrese la temperatura en n칰meros. Utilice el punto decimal si es necesario.")

#Funci칩n para convertir Celsius a Fahrenheit y Kelvin.
def convertir_celsius(TempCelsius):
    ResultFahrenheit = round((TempCelsius * (9/5)) + 32, 2)
    ResultKelvin = round(TempCelsius + 273.15, 2)
    print(f"Los {TempCelsius}춿C son equivalentes a {ResultFahrenheit}춿F y {ResultKelvin}K")

#Funci칩n para convertir Fahrenheit a Celsius y Kelvin.
def convertir_fahrenheit(TempFahrenheit):
    ResultCelsius = round((TempFahrenheit - 32) / (9/5), 2)
    ResultKelvin = round((TempFahrenheit - 32) / (9/5) + 273.15, 2)
    print(f"Los {TempFahrenheit}춿F son equivalentes a {ResultCelsius}춿C y {ResultKelvin}K")

#Funci칩n principal que maneja el flujo del programa.
def main():
    while True:
        conversor = elegir_conversor()
        
        if conversor == 1:
            TempCelsius = obtener_temperatura("Celsius (춿C)")
            convertir_celsius(TempCelsius)
        elif conversor == 2:
            TempFahrenheit = obtener_temperatura("Fahrenheit (춿F)")
            convertir_fahrenheit(TempFahrenheit)
        
        while True:
            reinicio = input("쯈uieres realizar otra conversi칩n de temperatura? (SI/NO): ").strip().upper()
            if reinicio == "SI":
                break
            elif reinicio == "NO":
                print("Gracias por usar el conversor de temperatura.")
                return
            else:
                print("Respuesta inv치lida. Por favor, ingresa 'SI' o 'NO'.")

if __name__ == "__main__":
    main()
```

### Ejercicio 3
#### Contador de palabras de un texto.

Escribe una funci칩n que dado un texto, muestre las veces que aparece cada palabra.

```python
#Funci칩n para obtener el texto ingresado por el usuario.
def obtener_texto():
    texto = input("Ingrese un texto y se analizar치 cu치ntas veces aparece cada palabra:\n")
    return texto

#Funci칩n para limpiar el texto: convertir a min칰sculas y eliminar signos de puntuaci칩n.
def limpiar_texto(texto):
    texto = texto.lower()  #M칠todo para convertir a min칰sculas
    texto_limpio = "" #Inicializa una variable vac칤a
    for caracter in texto:
        if caracter.isalnum() or caracter.isspace():  #Mantener caracteres alfanum칠ricos y espacios (Limpia caracteres especiales tipo ","".")
            texto_limpio += caracter
    return texto_limpio

#Funci칩n para contar las palabras en el texto y devolver un diccionario con los conteos.
def contar_palabras(texto):
    palabras = texto.split()  #Dividir el texto en palabras
    contador = {}  #Usar un diccionario est치ndar para contar las palabras
    for palabra in palabras:
        if palabra in contador:
            contador[palabra] += 1
        else:
            contador[palabra] = 1
    return contador

#Funci칩n para mostrar los resultados del conteo de palabras.
def mostrar_resultados(contador):
    print("\nResultados del an치lisis:")
    for palabra, cantidad in contador.items():
        print(f"{palabra}: {cantidad} veces")

#Funci칩n principal que maneja el flujo del programa.
def main():
    texto = obtener_texto()
    texto_limpio = limpiar_texto(texto)
    contador_palabras = contar_palabras(texto_limpio)
    mostrar_resultados(contador_palabras)

if __name__ == "__main__":
    main()
```

### Ejercicio 4
#### Diccionario inverso.

Resulta que el cliente tiene una encuesta muy antigua que se almacena en un diccionario y los resultados los necesita a la inversa, es decir, intercambiados las claves y valores. Los valores y claves en el diccionario original son 칰nicos; si 칠ste no es el caso, la funci칩n deber칤a imprimir un mensaje de advertencia.



```python
#Verificar si los valores en el diccionario son 칰nicos creando un set() (Un set no puede tener valores repetidos 游땔)
def intercambiar_claves_valores(diccionario):
    valores = list(diccionario.values())
    if len(valores) != len(set(valores)):
        print("Advertencia: Los valores en el diccionario no son 칰nicos.")
        return None
    #Crear un nuevo diccionario con claves y valores intercambiados
    diccionario_invertido = {v: k for k, v in diccionario.items()}
    return diccionario_invertido

#Ejemplo de uso
diccionario_original = {
    "nombre": "Fede",
    "edad": 29,
    "ciudad": "Barcelona",
    "deporte": "Nataci칩n"
}

#Ejecuto la funci칩n y si no son valores iguales inntercambio los valores.
diccionario_invertido = intercambiar_claves_valores(diccionario_original)

if diccionario_invertido:
    print("Diccionario original:", diccionario_original)
    print("Diccionario invertido:", diccionario_invertido)
```

### Ejercicio 1
#### Diccionario inverso con duplicados

Continuando con el ejercicio 4 del nivel 1: al cliente se olvid칩 comentar un detalle y resulta que los valores en el diccionario original pueden duplicarse y m치s, por lo que las claves intercambiadas pueden tener duplicados. En este caso, en el ejercicio anterior imprim칤as un mensaje de advertencia, ahora, los valores del diccionario resultante tendr치n que almacenarse como una lista. Ten en cuenta que si es un valor 칰nico no debe ser una lista.



```python
#Funci칩n para intercambiar claves y valores
def intercambiar_claves_valores(diccionario):
    diccionario_invertido = {} #Diccionario vac칤o para almacenar los valores

    for clave, valor in diccionario.items():
        if valor in diccionario_invertido: #Verifica si el valor que est치 iterando ya est치 en el diccionario_invertido
            if isinstance(diccionario_invertido[valor], list): #Si el valor es una lista: 游녢
                diccionario_invertido[valor].append(clave) #A침ane la clava en esa lista con .append
            else: #Si el valor no es una lista: 游녢
                diccionario_invertido[valor] = [diccionario_invertido[valor], clave] #Convierte el valor 칰nico en una lista que contiene tanto el valor original como la nueva clave
        else: #Si el valor no est치 en diccionario_invertido
            diccionario_invertido[valor] = clave #Lo a침ade como una nueva entrada en el diccionario_invertido con la clave actual
    
    return diccionario_invertido #La funci칩n devuelve el diccionario_invertido

#Ejemplo de uso
diccionario_original = {
    "nombre": "Fede",
    "edad": 29,
    "ciudad": "Barcelona",
    "deporte": "Nataci칩n",
    "hobby": "Teatro",
    "nombre del padre": "Fede",
    "veces que ejecut칩 este programa": 29
}

#Main del programa: Ejecutar la funci칩n y visualizar
diccionario_invertido = intercambiar_claves_valores(diccionario_original)
print("Diccionario original:", diccionario_original)
print("Diccionario invertido:", diccionario_invertido)
```
