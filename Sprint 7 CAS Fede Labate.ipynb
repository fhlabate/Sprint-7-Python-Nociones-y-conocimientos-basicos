### Ejercicio 1
#### Calculadora de l"índex de massa corporal
* Escribe una función que calcule el IMC ingresado por el usuario/a, es decir, quien lo ejecute deberá ingresar estos datos.
Puedes obtener más información de su cálculo en:
[Índice de masa corporal IMC que es y cómo se calcula](https://muysalud.com/salud/indice-de-masa-corporal-imc-que-es-y-como-se-calcula/)
* La función debe clasificar el resultado en sus respectivas categorías.

```python
#Función para que el usuario introduzca su peso.
def obtener_peso():
    while True:
        try:
            return float(input("Introduzca su peso en kilos (Ejemplo: 65.5): "))
        except ValueError:
            print("Por favor ingrese su peso en números, si tiene decimales utilice el punto")

#Función para que el usuario introduzca su estatura.
def obtener_estatura():
    while True:
        try:
            return float(input("Introduzca su estatura en metros (Ejemplo: 1.70): "))
        except ValueError:
            print("Por favor ingrese su estatura en números, si tiene decimales utilice el punto")

#Función para calcular el IMC.
def calcular_IMC():
    peso = obtener_peso() #Llamo a la función y me devuelve un valor de peso.
    estatura = obtener_estatura() #Llamo a la función y me devuelve un valor de estatura.
    IMC = round(peso / (estatura**2), 1) #Hago la operación para saber el IMC
    casos_de_IMC(IMC) #Llamo a la función y me devuelve el resultado que quiero mostrar.

#Función de mensajes de los 4 tipos de IMC.
def casos_de_IMC(IMC):
    if IMC < 18.5:
        print(f"Tu IMC es {IMC} e indica que estás bajo de peso")
    elif IMC >= 18.5 and IMC < 24.9:
        print(f"Tu IMC es {IMC} e indica que estás en un peso normal")
    elif IMC >= 24.9 and IMC < 29.9:
        print(f"Tu IMC es {IMC} e indica que tienes sobrepeso")
    else:
        print(f"Tu IMC es {IMC} e indica que tienes obesidad")

#Función principal que maneja el flujo del programa.
def main():
    while True:
        calcular_IMC()
        while True:
            reinicio = input("¿Quieres calcular otro IMC? (SI/NO): ").strip().upper()
            if reinicio == "SI":
                break
            elif reinicio == "NO":
                print("Gracias por usar el calculador de IMC")
                return
            else:
                print("Respuesta inválida. Por favor, ingresa 'SI'o 'NO'.")

if __name__ == "__main__":
    main()

#Cuando colocas if __name__ == "__main__": en tu archivo de Python, le estás diciendo a Python que ejecute el bloque de código que sigue (en este caso, main()) 
#solo si el archivo se está ejecutando como el script principal y no si se está importando como un módulo en otro archivo.
```

### Ejercicio 2
#### Convertidor de temperaturas.

Existen diversas unidades de temperatura utilizadas en distintos contextos y regiones. 
Las más comunes son:
* Celsius (°C)
* Fahrenheit (°F)
* Kelvin (K).

También existen otras unidades como:
* Rankine (°Ra) 
* Réaumur (°Re). 

Selecciona al menos 2 conversores, de modo que al introducir una temperatura devuelva, como mínimo, dos conversiones.

```python
#Función para que el usuario elija el tipo de conversor.
def elegir_conversor():
    print("Ingrese el tipo de conversor que desea utilizar:")
    while True:
        conversor = input("1-Celsius (°C) a Fahrenheit (°F) y Kelvin (K)\n"
                          "2-Fahrenheit (°F) a Celsius (°C) y Kelvin (K)\n")
        if conversor in ["1", "2"]:
            return int(conversor)
        else:
            print("Entrada incorrecta. Por favor, ingrese 1 o 2.")

#Función para obtener la temperatura del usuario.
def obtener_temperatura(tipo):
    while True:
        temp_str = input(f"Ingrese la temperatura en grados {tipo}: ")
        try:
            temp = float(temp_str)
            return temp
        except ValueError:
            print("Por favor, ingrese la temperatura en números. Utilice el punto decimal si es necesario.")

#Función para convertir Celsius a Fahrenheit y Kelvin.
def convertir_celsius(TempCelsius):
    ResultFahrenheit = round((TempCelsius * (9/5)) + 32, 2)
    ResultKelvin = round(TempCelsius + 273.15, 2)
    print(f"Los {TempCelsius}°C son equivalentes a {ResultFahrenheit}°F y {ResultKelvin}K")

#Función para convertir Fahrenheit a Celsius y Kelvin.
def convertir_fahrenheit(TempFahrenheit):
    ResultCelsius = round((TempFahrenheit - 32) / (9/5), 2)
    ResultKelvin = round((TempFahrenheit - 32) / (9/5) + 273.15, 2)
    print(f"Los {TempFahrenheit}°F son equivalentes a {ResultCelsius}°C y {ResultKelvin}K")

#Función principal que maneja el flujo del programa.
def main():
    while True:
        conversor = elegir_conversor()
        
        if conversor == 1:
            TempCelsius = obtener_temperatura("Celsius (°C)")
            convertir_celsius(TempCelsius)
        elif conversor == 2:
            TempFahrenheit = obtener_temperatura("Fahrenheit (°F)")
            convertir_fahrenheit(TempFahrenheit)
        
        while True:
            reinicio = input("¿Quieres realizar otra conversión de temperatura? (SI/NO): ").strip().upper()
            if reinicio == "SI":
                break
            elif reinicio == "NO":
                print("Gracias por usar el conversor de temperatura.")
                return
            else:
                print("Respuesta inválida. Por favor, ingresa 'SI' o 'NO'.")

if __name__ == "__main__":
    main()
```

### Ejercicio 3
#### Contador de palabras de un texto.

Escribe una función que dado un texto, muestre las veces que aparece cada palabra.

```python
#Función para obtener el texto ingresado por el usuario.
def obtener_texto():
    texto = input("Ingrese un texto y se analizará cuántas veces aparece cada palabra:\n")
    return texto

#Función para limpiar el texto: convertir a minúsculas y eliminar signos de puntuación.
def limpiar_texto(texto):
    texto = texto.lower()  #Método para convertir a minúsculas
    texto_limpio = "" #Inicializa una variable vacía
    for caracter in texto:
        if caracter.isalnum() or caracter.isspace():  #Mantener caracteres alfanuméricos y espacios (Limpia caracteres especiales tipo ","".")
            texto_limpio += caracter
    return texto_limpio

#Función para contar las palabras en el texto y devolver un diccionario con los conteos.
def contar_palabras(texto):
    palabras = texto.split()  #Dividir el texto en palabras
    contador = {}  #Usar un diccionario estándar para contar las palabras
    for palabra in palabras:
        if palabra in contador:
            contador[palabra] += 1
        else:
            contador[palabra] = 1
    return contador

#Función para mostrar los resultados del conteo de palabras.
def mostrar_resultados(contador):
    print("\nResultados del análisis:")
    for palabra, cantidad in contador.items():
        print(f"{palabra}: {cantidad} veces")

#Función principal que maneja el flujo del programa.
def main():
    texto = obtener_texto()
    texto_limpio = limpiar_texto(texto)
    contador_palabras = contar_palabras(texto_limpio)
    mostrar_resultados(contador_palabras)

if __name__ == "__main__":
    main()
```

### Ejercicio 4
#### Diccionario inverso.

Resulta que el cliente tiene una encuesta muy antigua que se almacena en un diccionario y los resultados los necesita a la inversa, es decir, intercambiados las claves y valores. Los valores y claves en el diccionario original son únicos; si éste no es el caso, la función debería imprimir un mensaje de advertencia.



```python
#Verificar si los valores en el diccionario son únicos creando un set() (Un set no puede tener valores repetidos 😉)
def intercambiar_claves_valores(diccionario):
    valores = list(diccionario.values())
    if len(valores) != len(set(valores)):
        print("Advertencia: Los valores en el diccionario no son únicos.")
        return None
    #Crear un nuevo diccionario con claves y valores intercambiados
    diccionario_invertido = {v: k for k, v in diccionario.items()}
    return diccionario_invertido

#Ejemplo de uso
diccionario_original = {
    "nombre": "Fede",
    "edad": 29,
    "ciudad": "Barcelona",
    "deporte": "Natación"
}

#Ejecuto la función y si no son valores iguales inntercambio los valores.
diccionario_invertido = intercambiar_claves_valores(diccionario_original)

if diccionario_invertido:
    print("Diccionario original:", diccionario_original)
    print("Diccionario invertido:", diccionario_invertido)
```

### Ejercicio 1
#### Diccionario inverso con duplicados

Continuando con el ejercicio 4 del nivel 1: al cliente se olvidó comentar un detalle y resulta que los valores en el diccionario original pueden duplicarse y más, por lo que las claves intercambiadas pueden tener duplicados. En este caso, en el ejercicio anterior imprimías un mensaje de advertencia, ahora, los valores del diccionario resultante tendrán que almacenarse como una lista. Ten en cuenta que si es un valor único no debe ser una lista.



```python
#Función para intercambiar claves y valores
def intercambiar_claves_valores(diccionario):
    diccionario_invertido = {} #Diccionario vacío para almacenar los valores

    for clave, valor in diccionario.items():
        if valor in diccionario_invertido: #Verifica si el valor que está iterando ya está en el diccionario_invertido
            if isinstance(diccionario_invertido[valor], list): #Si el valor es una lista: 👇
                diccionario_invertido[valor].append(clave) #Añane la clava en esa lista con .append
            else: #Si el valor no es una lista: 👇
                diccionario_invertido[valor] = [diccionario_invertido[valor], clave] #Convierte el valor único en una lista que contiene tanto el valor original como la nueva clave
        else: #Si el valor no está en diccionario_invertido
            diccionario_invertido[valor] = clave #Lo añade como una nueva entrada en el diccionario_invertido con la clave actual
    
    return diccionario_invertido #La función devuelve el diccionario_invertido

#Ejemplo de uso
diccionario_original = {
    "nombre": "Fede",
    "edad": 29,
    "ciudad": "Barcelona",
    "deporte": "Natación",
    "hobby": "Teatro",
    "nombre del padre": "Fede",
    "veces que ejecutó este programa": 29
}

#Main del programa: Ejecutar la función y visualizar
diccionario_invertido = intercambiar_claves_valores(diccionario_original)
print("Diccionario original:", diccionario_original)
print("Diccionario invertido:", diccionario_invertido)
```
